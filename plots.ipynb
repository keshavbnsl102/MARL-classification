```python
import json
import matplotlib.pyplot as plt

two_agent = "outputs/6_5.json"
four_agent = "outputs/4_5.json"


with open(two_agent, "r") as read_file:
    data_2 = json.load(read_file)

with open(four_agent, "r") as read_file:
    data_4 = json.load(read_file)


a_2 = list(data_2.values())
a_4 = list(data_4.values())


plt.plot(a_2)
plt.plot(a_4)
plt.xlabel("epochs")
plt.ylabel("error")
plt.title("#frame - 4 v.s. 6");
```
